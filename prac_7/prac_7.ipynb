```python
'''7(b) Write a program to display a line graph with x = [1, 2, 3, 4] and y = [1, 4, 9, 16]. 
Give appropriate labels to both the axes.'''
import matplotlib.pyplot as plt
%matplotlib inline
values1 = [1, 2, 3, 4]
values2 = [1, 4, 9, 16]

plt.plot(values1)
plt.plot(values2)

plt.xlabel('X-Values')
plt.ylabel('Y-Values')

plt.legend(['CX','CY'],loc=4)
plt.show()
```

```python
'''7(c) Write a program to display a bar chart where x-axis represents student enrolment numbers and y-axis represents
CPI. Give appropriate labels to both the axes.'''
import matplotlib.pyplot as plt
%matplotlib inline
x = [1,2,3,4,5]
y = [5.9,6.2,3.2,8.9,9.7]
c = ['b','g','r','c','m']
w = [0.5,0.6,0.3,0.8,0.9]
plt.title('Student\'s CPI')
plt.xlabel('Enrolment numbers')
plt.ylabel('CPI')
plt.bar(x,y,color=c,width=w)
plt.show()
```
